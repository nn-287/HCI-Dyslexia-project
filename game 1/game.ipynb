```python
import math
import cv2
import random
import cvzone
import numpy as np

class SnakeGameClass:
    def __init__(self,pathFood):
        self.initializeData()
        self.gameOver=False
        self.score=1
        # interactging with the food
        self.imgFood=cv2.imread(str(lol)+".png",cv2.IMREAD_UNCHANGED)
        self.hFood,self.wFood , _ =self.imgFood.shape 
        self.imgFake=cv2.imread(str(fake)+".png",cv2.IMREAD_UNCHANGED)
        self.hFake,self.wFake , _ =self.imgFake.shape 
        
    def randomLocation(self):
        self.foodPoint=random.randint(100,500),random.randint(100,300)
        self.fakePoint=random.randint(600,800),random.randint(100,300)
    def initializeData(self):
        
        self.previousHead=0,0
        self.foodPoint=0,0
        self.fakePoint=0,0
        self.randomLocation()
        
        
        
    def update(self,imgMain,currentHead):
        global lol
        global fake
        if self.gameOver:
            cvzone.putTextRect(imgMain,"Game Over",[300//2,400//2],scale=3,thickness=3,
                            offset=20)
            cvzone.putTextRect(imgMain,f'Your Score {self.score}',[300//2,550//2],scale=3,thickness=3,
                            offset=20)
            
        else:
            
            cx,cy=currentHead
            px,py=self.previousHead
         
            distance= math.hypot(cx-px,cy-py)
            
            rx,ry=self.foodPoint
            if(rx-self.wFood//2 <cx<rx+self.wFood//2 and ry-self.hFood//2
               <cy<ry+self.hFood//2):
                self.score+=1
                lol+=1
                fake-=1
                if(self.score%23==0):
                    lol=1
                    fake=23

                self.imgFood=cv2.imread(str(lol)+".png",cv2.IMREAD_UNCHANGED)
                self.imgFake=cv2.imread(str(fake)+".png",cv2.IMREAD_UNCHANGED)
                self.hFood,self.wFood , _ =self.imgFood.shape
                self.hFake,self.wFake , _ =self.imgFake.shape
              
                self.randomLocation()
            #
            fx,fy=self.fakePoint
            if(fx-self.wFake//2 <cx<fx+self.wFake//2 and fy-self.hFake//2
               <cy<fy+self.hFake//2):
                self.gameOver=True
                    

            
            #draw letter
            cvzone.putTextRect(imgMain,f'Score {self.score}',[50,80],scale=3,thickness=2,
                                offset=10)
            cvzone.putTextRect(imgMain,f'coor {str(rx)+"   "+str(ry)}',[500,80],scale=3,thickness=2,
                                offset=10)
            imgMain=cvzone.overlayPNG(imgMain,self.imgFood, (rx-self.wFood//2,ry-self.hFood//2))
            
            imgMain=cvzone.overlayPNG(imgMain,self.imgFake, (fx-self.wFake//2,fy-self.hFake//2))
            
           
        return imgMain
                
   


from cvzone.HandTrackingModule import HandDetector

cap=cv2.VideoCapture(0);
cap.set(3,1280) #the size frame
cap.set(4,1024)
detector=HandDetector(detectionCon=0.8,maxHands=1)
lol=1
fake=23
game=SnakeGameClass(str(lol)+".png")

while True:
    success , img=cap.read();
    img=cv2.flip(img,1)
    hands,img=detector.findHands(img,flipType =False)
    if hands:# if i detect at least one hand
        lmList=hands[0]["lmList"]
        pointIndex=lmList[8][0:2]
        img=game.update(img,pointIndex)
    cv2.imshow("img", img)
    
    if(cv2.waitKey(1)== ord('r')):
        game.gameOver=False
        game.score=1
    
    if  (cv2.waitKey(1)== ord('q')):
        break
    if(lol==23):
        lol=1
        fake=23
```
